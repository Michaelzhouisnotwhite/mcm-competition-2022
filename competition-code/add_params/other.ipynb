```python
import numpy as np
import pandas as pd
```

```python
gen_df = pd.read_csv("../general_table.csv")
gen_df.head()
```

```python
gen_df['gold_ma5'] = gen_df[gen_df.columns[1]].rolling(5).mean()
gen_df['bit_ma5'] = gen_df[gen_df.columns[2]].rolling(5).mean()
gen_df['gold_ma15'] = gen_df[gen_df.columns[1]].rolling(15).mean()
gen_df['bit_ma15'] = gen_df[gen_df.columns[2]].rolling(15).mean()

gen_df.head()
```

```python
gen_df['gold_ma5_bias'] = 100 *(gen_df['USD (PM)'] - gen_df['gold_ma5']) / gen_df['gold_ma5']
gen_df['gold_ma15_bias'] = 100 *(gen_df['USD (PM)'] - gen_df['gold_ma15']) / gen_df['gold_ma15']

gen_df['bit_ma5_bias'] = 100 * (gen_df['Value'] - gen_df['bit_ma5']) / gen_df['bit_ma5']
gen_df['bit_ma15_bias'] = 100 * (gen_df['Value'] - gen_df['bit_ma15']) / gen_df['bit_ma15']

gen_df.head()
```

```python
gen_df.columns
```

```python
gen_df['gold_growth_rate'] = gen_df['gold_growth_rate'] * 100
gen_df['bit_growth_rate'] = gen_df['bit_growth_rate'] * 100

gen_df['gold_quantile'] = gen_df['gold_quantile'] * 100
gen_df['bit_quantile'] = gen_df['bit_quantile'] * 100
gen_df.to_csv("../bias-means-table.csv")
```

```python
# 黄金15日震荡
# 比特币5日震荡
def get_mean_growth_positive(rb, n):
    return rb[rb > 0].sum() / n

def get_mean_growth_negative(rb, n):
    return  - rb[rb < 0].sum() / n

def get_rsi(positive, negative):
    return (positive / (positive + negative)) * 100    

pg = gen_df['gold_growth_rate'].rolling(15).apply(get_mean_growth_positive, args=(15, ))
ng = gen_df['gold_growth_rate'].rolling(15).apply(get_mean_growth_negative, args=(15, ))

pb = gen_df['bit_growth_rate'].rolling(5).apply(get_mean_growth_positive, args=(5, ))
nb = gen_df['bit_growth_rate'].rolling(5).apply(get_mean_growth_negative, args=(5, ))

gen_df['gold_rsi_15'] = get_rsi(pg, ng)
gen_df['bit_rsi_5'] = get_rsi(pb, nb)
```

```python
gen_df['gold_growth'] = 0
gen_df['bit_growth'] = 0
for i in gen_df.index[1:]:
    gen_df['gold_growth'].iloc[i] = gen_df['USD (PM)'].iloc[i] - gen_df['USD (PM)'].iloc[i - 1]
    gen_df['bit_growth'].iloc[i] = gen_df['Value'].iloc[i] - gen_df['Value'].iloc[i - 1]
    
```

```python
# 盈亏比
gen_df['gold_pl_rate'] = 0
gen_df['bit_pl_rate'] = 0
for i in gen_df.index[1:]:
    prev_data_gold = gen_df['gold_growth'].iloc[:i + 1]
    gen_df['gold_pl_rate'].iloc[i] = prev_data_gold[prev_data_gold >
                                                    0].mean() / prev_data_gold[prev_data_gold < 0].mean()

    prev_data_bit = gen_df['bit_growth'].iloc[:i + 1]
    gen_df['bit_pl_rate'].iloc[i] = prev_data_bit[prev_data_bit >
                                                    0].mean() / prev_data_bit[prev_data_bit < 0].mean()
```
