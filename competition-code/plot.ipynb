```python
import numpy as np
import pandas as pd
import pylab as plt
import matplotlib
```

```python
matplotlib.rcParams['font.family'] = 'SimHei'
plt.rcParams['axes.unicode_minus'] = False
```

```python
gen_data = pd.read_csv('盈亏比.csv', index_col=0)
```

```python
fig = plt.figure(figsize=(50, 10))
ax1 = fig.add_subplot(211)
i = 0
plt.title("gold_macd", fontdict={'fontsize': 20})
gen_data['gold_macd'].iloc[i:i+365].plot(ax=ax1, kind='bar', color='orangered')
gen_data['gold_dif'].iloc[i:i+365].plot(ax=ax1, kind='line', xticks=range(i, i+365, 30))
gen_data['gold_dea'].iloc[i:i+365].plot(ax=ax1, kind='line', xticks=range(i, i+365, 30))
ax1.set_xticklabels(labels=range(i, i+365, 30), rotation=0, fontsize=20)
for label in ax1.get_yticklabels():
    label.set_fontsize(20)
ax2 = fig.add_subplot(212)

gen_data['gold_macd'].iloc[1465:].plot(ax=ax2, kind='bar', color='orangered', xticks=range(i, i+365, 30))
ax2.plot(gen_data['gold_dif'].iloc[1465:].values, ls='-', label="gold_dif")
ax2.plot(gen_data['gold_dea'].iloc[1465:].values, ls='-', label="gold_dea")

for label in ax2.get_yticklabels():
    label.set_fontsize(20)
ax2.set_xticklabels(labels=range(i, i+365, 30), rotation=0, fontsize=20)
plt.legend(fontsize=20)

plt.savefig('第一年和最后一年的黄金macd图.png')
plt.show()

fig = plt.figure(figsize=(50, 10))
ax1 = fig.add_subplot(211)
i = 0
plt.title("bitcorn_macd", fontdict={'fontsize': 20})

gen_data['bit_macd'].iloc[i:i+365].plot(ax=ax1, kind='bar', color='purple')
gen_data['bit_dif'].iloc[i:i+365].plot(ax=ax1, kind='line', xticks=range(i, i+365, 30), color='blue')
gen_data['bit_dea'].iloc[i:i+365].plot(ax=ax1, kind='line', xticks=range(i, i+365, 30), color="green")
ax1.set_xticklabels(labels=range(i, i+365, 30), rotation=0, fontsize=20)
for label in ax1.get_yticklabels():
    label.set_fontsize(20)
ax2 = fig.add_subplot(212)

gen_data['bit_macd'].iloc[1465:].plot(ax=ax2, kind='bar', color='purple', xticks=range(i, i+365, 30))
ax2.plot(gen_data['bit_dif'].iloc[1465:].values, ls='-', label="gold_dif", color='blue')
ax2.plot(gen_data['bit_dea'].iloc[1465:].values, ls='-', label="gold_dea", color='green')

for label in ax2.get_yticklabels():
    label.set_fontsize(20)
ax2.set_xticklabels(labels=range(i, i+365, 30), rotation=0, fontsize=20)
plt.legend(fontsize=20)

plt.savefig('第一年和最后一年的比特币macd图.png')
plt.show()
```

```python
fig = plt.figure(figsize=(20, 10))
ax1.set_title("Gold MA", fontdict={"fontsize":20})
ax1, ax2 = fig.add_subplot(211), fig.add_subplot(212)
gen_data[['USD (PM)', 'gold_ma5', 'gold_ma15']].loc[:365].plot(ax=ax1, xticks=range(0, 360, 30))
gen_data[['USD (PM)', 'gold_ma5', 'gold_ma15']].loc[1465:].plot(ax=ax2)
plt.savefig('第一年和最后一年黄金ma.png')
plt.show()
fig = plt.figure(figsize=(20, 10))
ax1.set_title("Bitcorn MA", fontdict={"fontsize":20})
ax1, ax2 = fig.add_subplot(211), fig.add_subplot(212)
gen_data[['Value', 'bit_ma5', 'bit_ma15']].loc[:365].plot(ax=ax1, xticks=range(0, 360, 30))
gen_data[['Value', 'bit_ma5', 'bit_ma15']].loc[1465:].plot(ax=ax2)
plt.savefig('第一年和最后一年比特币ma.png')

plt.show()
```
